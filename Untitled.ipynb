```python
import pandas as pd
import numpy as np
```

```python
spath = 'Z:\\data_XAS\\Tourmalines\\Fe\\Fe_tourmaline_standards_spectra_081823.csv'
spectra = pd.read_csv(spath)
mpath = 'Z:\\data_XAS\\Tourmalines\\Fe\\Fe_tourmaline_standards_metadata_081823.csv'
meta = pd.read_csv(mpath)
```

```python
for sample in meta.Sample.unique():
    for a in ['no','yes']:
        pkeys = list((meta[
                      (meta.Sample==sample)&
                      (meta.Quality=='good')&
                      (meta['Absorption Corrected?']==a)])['pkey'].values)
        if len(pkeys) > 0:
            mean_spectrum = list(spectra[pkeys].mean(axis=1))
            new_sample = sample.replace(' ','_')
            new_sample = f'AVERAGE_{new_sample}_{a}_abscorr'
            spectra[new_sample] = mean_spectrum
```
